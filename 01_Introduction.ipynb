```python
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

### Define functions

```python
def create_1d_data(num_examples, w, b, bound):
  """Create X, Y data with a linear relationship with added noise.

  Args:
    num_examples: number of examples to generate
    w: desired slope
    b: desired intercept
    bound: lower and upper boundary of the data interval

  Returns:
    X and Y with shape (num_examples)
  """
  np.random.seed(4)  # consistent random number generation
  X = np.arange(num_examples)
  deltas = np.random.uniform(low=-bound, high=bound, size=X.shape) # added noise
  Y = b + deltas + w * X

  return X, Y
```

### Step 5: Evaluation and Generalization

How good are our models? Intuitively, the better the model, the more closely it fits the data we have. That is, for each $x$, we'll compare $y$, the true value, with $\hat{y}$, the predicted value. This comparison is often called the *loss* or the *error*. One common such comparison is *squared error*: $(y-\hat{y})^2$. Averaging over all our data points, we get the *mean squared error*:

\begin{equation}
\textit{MSE} = \frac{1}{n} \sum_{y_i \in Y}(y_i - \hat{y}_i)^2
\end{equation}

How well do our models generalize? The test dataset serves as a proxy for unseen data in real-world applications. By evaluating the model on the test data, you can assess its ability to generalize beyond the training data. This ensures that the model can make accurate predictions on new data it hasn't seen during training.

### <span style="color:chocolate">Exercise 5:</span> Computing MSE (20 points)

```python
# YOUR CODE HERE
def MSE(true_values, predicted_values):
  """Return the MSE between true_values and predicted values."""
  return np.mean((true_values - predicted_values) ** 2)
```

#### <span style="color:chocolate">Tensorflow</span>
To implement this project using TensorFlow, we would need to modify the code to leverage TensorFlow's API for creating models, computing predictions, and calculating loss functions like Mean Squared Error (MSE). Instead of using simple Python functions for the models, we'll define them as TensorFlow models.

Here's how you can adapt the code using TensorFlow:

Steps to Implement:
Data Generation: This part stays mostly the same, as we can use NumPy to create the dataset.
Splitting the Data: We'll continue using train_test_split from scikit-learn.
Model Definition: Instead of manually defining the models M1 and M2, we will use TensorFlow's Sequential model.
Training: TensorFlow's compile() method will be used to define the loss function, and fit() will train the models.
Evaluation: We'll use the evaluate() method for MSE calculations.

```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 1. Create the dataset
def create_1d_data(num_examples, w, b, bound):
    np.random.seed(4)  # Consistent random number generation
    X = np.arange(num_examples)
    deltas = np.random.uniform(low=-bound, high=bound, size=X.shape)  # Added noise
    Y = w * X + b + deltas
    return X, Y

X, Y = create_1d_data(num_examples=70, w=2, b=1, bound=1)

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)

# 2. Define the models using TensorFlow (M1, M2)
def build_linear_model(w, b):
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(1, input_shape=(1,), kernel_initializer=tf.constant_initializer(w), bias_initializer=tf.constant_initializer(b))
    ])
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Model M1: y = 5 + x
model_M1 = build_linear_model(1, 5)

# Model M2: y = 1 + 2x
model_M2 = build_linear_model(2, 1)

# Reshape the data to fit TensorFlow's input requirements (1D input)
X_train_tf = X_train.reshape(-1, 1)
Y_train_tf = Y_train.reshape(-1, 1)
X_test_tf = X_test.reshape(-1, 1)
Y_test_tf = Y_test.reshape(-1, 1)

# Train the models
model_M1.fit(X_train_tf, Y_train_tf, epochs=100, verbose=0)
model_M2.fit(X_train_tf, Y_train_tf, epochs=100, verbose=0)

# 3. Make predictions and plot results
M1_hat_train = model_M1.predict(X_train_tf)
M2_hat_train = model_M2.predict(X_train_tf)

plt.scatter(X_train, Y_train, label='Observed training data', color='blue')
plt.plot(X_train, M1_hat_train, label='M1(x) = 5 + x', color='green')
plt.plot(X_train, M2_hat_train, label='M2(x) = 1 + 2x', color='orange')
plt.xlabel('X_train')
plt.ylabel('Y_train')
plt.title('Model Predictions vs Observed Data')
plt.legend()
plt.show()

# 4. Compute MSE for both models on training and test sets
def compute_mse(model, X, Y):
    predictions = model.predict(X)
    mse = tf.reduce_mean(tf.square(Y - predictions)).numpy()
    return mse

M1_train_mse = compute_mse(model_M1, X_train_tf, Y_train_tf)
M2_train_mse = compute_mse(model_M2, X_train_tf, Y_train_tf)
M1_test_mse = compute_mse(model_M1, X_test_tf, Y_test_tf)
M2_test_mse = compute_mse(model_M2, X_test_tf, Y_test_tf)

print(f"MSE for M1 on training data: {M1_train_mse}")
print(f"MSE for M2 on training data: {M2_train_mse}")
print(f"MSE for M1 on test data: {M1_test_mse}")
print(f"MSE for M2 on test data: {M2_test_mse}")

# 5. Fit an 8th-degree polynomial using TensorFlow
# In TensorFlow, we typically don't fit higher-degree polynomials directly like this, 
# but we can use a Dense layer with polynomial features for comparison.
poly_coeffs = np.polyfit(X_train, Y_train, 8)
M3_hat_train = np.polyval(poly_coeffs, X_train)

plt.scatter(X_train, Y_train, label='Observed training data', color='blue')
plt.plot(X_train, M3_hat_train, label='M3 (8th-degree polynomial)', color='red', linestyle='--')
plt.xlabel('X_train')
plt.ylabel('Y_train')
plt.title('Model M3 Predictions vs Observed Data')
plt.legend()
plt.show()

# Calculate the training and test MSE for M3
M3_hat_test = np.polyval(poly_coeffs, X_test)
M3_train_mse = MSE(Y_train, M3_hat_train)
M3_test_mse = MSE(Y_test, M3_hat_test)

print(f"M3 Training MSE: {M3_train_mse}")
print(f"M3 Test MSE: {M3_test_mse}")
```

This code defines a function build_linear_model(w, b) that creates a linear regression model using TensorFlow's Sequential API. 

Heres what each part of the code does:
tf.keras.Sequential():

The Sequential class allows you to stack layers to create a neural network model. In this case, its used to define a simple linear model with just a single Dense layer.
tf.keras.layers.Dense(1):

The Dense layer represents a fully connected layer in a neural network.
1: This specifies that the layer has a single output (one prediction value for each input X).
Activation: Since linear regression doesnt need a non-linear activation function, it defaults to a linear activation.
input_shape=(1,):

This tells TensorFlow that each input to the model will be a 1-dimensional value (i.e., a single feature for each input, since you're working with 1D data).
kernel_initializer=tf.constant_initializer(w):

The kernel represents the weight of the model, which is set to a constant value w. In this case, the model is manually initialized to have the slope provided by w (as defined by the equations in the problem).
bias_initializer=tf.constant_initializer(b):

This sets the bias (intercept) of the model to a constant value b.
model.compile():

Optimizer: The model is compiled using the Adam optimizer, which is a popular choice for many optimization tasks in neural networks.
Loss: The loss function is set to mean_squared_error, which is standard for regression tasks.

Why Adam Was Chosen Here:
In the context of your project, Adam was chosen because:
1. You are training simple linear models, and Adam ensures efficient convergence even in small models.
2. It adapts learning rates automatically, meaning you dont need to fine-tune the learning rate manually, which is helpful when you're iterating and testing models quickly.
3. It's a widely used and well-proven optimizer that works well for both simple and complex problems, reducing the need to experiment with other optimizers like SGD.

1. Reshaping the Data:

Why?: TensorFlow expects input data to be in a 2D array shape. Even though you're working with 1D data (X_train and Y_train), you need to reshape the data to explicitly have 2 dimensions (i.e., rows and columns) where each sample is a row and each feature is a column.
X_train.reshape(-1, 1):
The -1 tells NumPy to automatically infer the number of rows based on the size of the data (so it keeps all data points).
The 1 indicates that there is 1 feature (i.e., each data point is a single value, because you are working with 1D data).

This turns a 1D array like [1, 2, 3] into a 2D array like [[1], [2], [3]].

2. Training the Models:

Trains the linear model M1 using the training data X_train_tf and Y_train_tf.
epochs=100: This means the model will go through the entire training dataset 100 times to adjust the weights and biases.
verbose=0: This suppresses the training output, so it runs silently.
model_M2.fit():
Trains the linear model M2 using the same training data. The training process is the same as for M1.
